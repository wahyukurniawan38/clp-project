```python
import hashlib
import numpy as np

class OB:
    def __init__(self):
        self.x = np.zeros([5,])
        self.y = np.zeros([1,3])
    def __hash__(self)->int:
        x_hash = hash(self.x.tobytes())
        y_hash = hash(self.y.tobytes())
        return hash(x_hash+y_hash)
    def __eq__(self, other: object) -> bool:
        x_is_same = np.allclose(self.x, other.x)
        y_is_same = np.allclose(self.y, other.y)
        return x_is_same and y_is_same
        
a = OB()
b = OB()
b.y[0,:] = np.asanyarray([0,0,0])
ob_list = [a]
ob_dict = {}
ob_dict[a] = True
if not b in ob_dict:
    ob_list += [b]
    ob_dict[b]=True
print(ob_list)
# ob_set = set([a,b])
```
